### mode 字串
|字元|說明|
|:-:|:-|
|r|讀取模式|
|w|寫入模式|
|x|檔案不存在時才寫入|
|a|附加模式|
|b|二進位模式|
|t|文字模式|
|+|更新模式|

```python
# 讀取檔案：
file = open('test.txt', 'r')
# 把所有資料一次全部讀進來
text = file.read()

print(text)
# 關閉檔案
file.close()
```

```python
# 讀取檔案：
# 如果沒有第二個參數，表示預設的模式是 'r'
file = open('test.txt')

# 一次全部讀進來，但是分行
text = file.readlines()

print(text)

# 關閉檔案
file.close()
```

```python
# 讀取檔案：
file = open('test.txt')

# 將資料一行一行的讀進來處理
for line in file:
     print(line)

# 關閉檔案
file.close()
```

```python
# 寫入檔案
file = open('test.txt','w')
file.write('this is a test!')
file.close()
```

```python
# 附加資料 (Append)
file = open('test.txt', 'a')
file.write('this is a test!')
file.close()
```

```python
# 如果不想每一次都要呼叫 close() 來關閉檔案，則可以使用 with … as 語法。
with open('test.txt') as file:
    text = file.read()
    print(text)
```
